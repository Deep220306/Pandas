```python
import numpy as np
import pandas as pd
```

Creating Data Frames

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
```

```python
df2 = pd.DataFrame(data_list)
df2
```

```python
cols = ['Name', 'Age', 'City', 'Salary']
df2 = pd.DataFrame(data_list, columns=cols)
df2
```

Selection and Indexing of Columns

Coloumn Selection

```python
df2 ["Name"]
```

```python
df2[["Name", "City"]]
```

Column Insertion

```python
df2["Designation"] = ['Engineer', 'Manager', 'Analyst', 'Director']
df2
```

```python
df2.drop("Designation", axis=1) # This will return a new DataFrame without the "Designation" column, but it won't modify df2 unless you set inplace=True.
```

Coloumn removal

```python
df2.drop("Designation", axis=1, inplace=True)
df2
```

```python
df2.drop(["Age", "Salary"], axis=1)
```

Conditional Statement

```python
df2[df2["Age"] > 30] # Filter rows where Age is greater than 30
```

```python
df2[(df2["Age"] > 30) & (df2["Salary"] > 62000)] #we need to use parentheses to group the conditions when using logical operators like & (and) or | (or) in pandas.
```
